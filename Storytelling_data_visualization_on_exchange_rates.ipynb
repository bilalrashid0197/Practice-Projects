### Introducing the Dataset

```python
# Importing the different libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
%matplotlib inline
```

```python
# Reading in the Dataset
exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
# Printing the first five rows of the dataset
exchange_rates.head()
```

```python
# Printing the last five rows of the dataset
exchange_rates.tail()
```

```python
# Getting an overview of the entire dataset
exchange_rates.info()
```

In total, our dataset has 41 columns and has 5699 rows. Some of the columns exist of null values some of those columns are `[Brazilian real ]`, `[Bulgarian lev ] `, and many more. In total 17 columns have null values. Out of the 41 columns, 38 of those columns are of datatype object and the other three are of datatype float64.

### Data Cleaning

```python
# We rename the [US dollar ] and Period\Unit: columns to something easier to type like US_dollar and Time.
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar','Period\\Unit:': 'Time'},inplace=True)
# Changing the Time column to a datetime data type.
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
# Sorting the values by Time in ascending order.
exchange_rates.sort_values('Time', inplace=True)
# Reseting the index (and drop the initial index).
exchange_rates.reset_index(drop=True, inplace=True)
# Printing the first five rows
exchange_rates.head()
```

```python
# Isolating the Time and the US_dollar columns, we do this since our main focus in this project is on the rate between the Euro and the US dollar
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
# Checking the US_dollar column first 5 rows
print(euro_to_dollar.loc[:,"US_dollar"].value_counts().head())
print("\n")
# Checking the US_dollar columns last 5 rows
print(euro_to_dollar.loc[:,"US_dollar"].value_counts().tail())
print("\n")
# Exploring the new dataframe
euro_to_dollar.info()
```

After exploring the US_dollar column more we see 62 times the value "-" appears.  Since we don't want this type of data we will do the following:

- First, we will remove the "-" object out of the US_dollar columns
- Next, we will change the datatype form object to float64 of the US_dollar columns 

Below we perform these steps.

```python
# Removing "-" from the US_dollar column
euro_to_dollar = euro_to_dollar.loc[euro_to_dollar.loc[:,"US_dollar"] != "-"]
# Chagning the US_dollar from an object datatype to float64 datatype
euro_to_dollar.loc[:,"US_dollar"] = euro_to_dollar.loc[:,"US_dollar"].astype(float)
# Exploring our dataframe
euro_to_dollar.info()
```

### Rolling Mean

Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
# Poltting a line plot 
plt.plot(euro_to_dollar.loc[:,"Time"],euro_to_dollar.loc[:,"US_dollar"])
plt.show()
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run 
(months or years).

If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

```python
# Creating the figure
plt.figure(figsize=(9,6))
# Plotting the graph on the first index
plt.subplot(3,2,1)
# Ploting the line plot
plt.plot(euro_to_dollar.loc[:,"Time"],euro_to_dollar.loc[:,"US_dollar"])
# Adding a title
plt.title("Original Values",weight = "bold")

# Using a for loop to plot the other five graph
for i,rolling_mean in zip([2, 3, 4, 5, 6],[7, 30, 50, 100, 365]):
    # i sets the location of the plot(the index)
    plt.subplot(3,2,i)
    # rolling_mean takes a postion of the rolling_mean, for the 1st itteration it is 7, next 30 and etc.
    plt.plot(euro_to_dollar.loc[:,"Time"],euro_to_dollar.loc[:,"US_dollar"].rolling(rolling_mean).mean())
    # Adding a title
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')

plt.tight_layout() # Adjust the padding between and around subplots. 
plt.show()    
```

```python
# Creating a new columns and assging it to the rolling mean of every 30 days
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
# Viewing the dataframe
euro_to_dollar.info()
```

## Financial Crisis Graph

Below we are going to plot this graph:
   
   - We show how the euro-dollar rate changed during the 2007-2008's financial crisis. We can also show the data for 2016 and 2009 for comparison. 

#### Filtering the data

```python
financial_crisis_years_6_9 = euro_to_dollar.copy()[(euro_to_dollar.loc[:,"Time"].dt.year >= 2006) & (euro_to_dollar.loc[:,'Time'].dt.year <= 2009)]
financial_crisis_years_7_8 = euro_to_dollar.copy()[(euro_to_dollar.loc[:,"Time"].dt.year >= 2007) & (euro_to_dollar.loc[:,"Time"].dt.year <=2008)]
financial_crisis_years_6_9.info()
```

#### Plotting the line plot

```python
# Adding the FiveThirtyEight style
style.use("fivethirtyeight")
fig, ax1 = plt.subplots(figsize = (8,3))
axes = [ax1]
# Adding the plot
ax1.plot(financial_crisis_years_6_9.loc[:,"Time"],financial_crisis_years_6_9.loc[:,"rolling_mean"], color = "green", linewidth = 1)
# Highlighting the 2007-2008 period
ax1.plot(financial_crisis_years_7_8.loc[:,"Time"],financial_crisis_years_7_8.loc[:,"rolling_mean"],color ="red", linewidth = 2)
# Highlihting the peak of the crisis
ax1.axvspan(xmin=13950, xmax=14140, ymin=0.09,
           alpha=0.3, color='grey')
#Adding separate tick labels
for axes_object in axes:
    axes_object.set_yticklabels([])
    axes_object.set_xticklabels([])
# Setting the x-tick labels
x = 13110
for year in ["2006","2007","2008","2009","2010"]:
    ax1.text(x = x, y = 1.13 , s = year, alpha = 0.5 , fontsize = 11)
    x+=365
# Setting the y-tick labels
y = 1.193
for rate in ["1.2","1.3","1.4","1.5"]:
    ax1.text(x = 13020, y = y, s = rate, alpha = 0.5, fontsize =11)
    y+=0.1
# Adding a title and a subtitle
for axes_object in axes:
    axes_object.text(x = 13020, y = 1.67, s = "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",weight = "bold")
    axes_object.text(x = 13020, y = 1.63, s = "Euro-USD exchange rates between 2006 and 2010")
# Copyright
ax1.text(x = 13020, y = 1.07, s = "©Bilal Rashid" + " "*50 +"Source: European Central Bank",color = '#f0f0f0', backgroundcolor = '#4d4d4d')
plt.show()
```

The global financial crisis (GFC) of 2007–2008 was a severe global economic downturn. The United States, suffered when the value of their currency in relation to the euro reached an all-time high. During the financial crisis, the United States lost $7.4 trillion in stock wealth and $5.5 million more American jobs as a result of slower economic growth, resulting in one euro equaling 1.59 US dollars.

## Euro-USD Rate Under Last Three Us Presidents

Below we are going to plot this graph:

    - We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

#### Filtering the data

```python
george_bush = euro_to_dollar.copy()[(euro_to_dollar.loc[:,"Time"].dt.year >= 2001) & (euro_to_dollar.loc[:,"Time"].dt.year < 2009)]
barack_obama = euro_to_dollar.copy()[(euro_to_dollar.loc[:,"Time"].dt.year >= 2009) & (euro_to_dollar.loc[:,"Time"].dt.year < 2017)]
donald_trump = euro_to_dollar.copy()[(euro_to_dollar.loc[:,"Time"].dt.year >= 2017) & (euro_to_dollar.loc[:,"Time"].dt.year < 2021)]
```

#### Plotting the line plot

```python
# Adding the style we will be using
style.use("fivethirtyeight")
# Creating the different plot we will be using
plt.figure(figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1,ax2,ax3,ax4]

# Plotting the different plot of the different presidents
### Plot of George Bush
ax1.plot(george_bush.loc[:,"Time"], george_bush.loc[:,"rolling_mean"], color = '#BF5FFF')
### Plot of Barak Obama
ax2.plot(barack_obama.loc[:,"Time"], barack_obama.loc[:,"rolling_mean"], color = "#ffa500")
### Plot of Donald Trump
ax3.plot(donald_trump.loc[:,"Time"], donald_trump.loc[:,"rolling_mean"], color = "#00B2EE")
### Plotting the overall rate of change over the three periods of presidency
ax4.plot(george_bush.loc[:,"Time"], george_bush.loc[:,"rolling_mean"], color = '#BF5FFF')
ax4.plot(barack_obama.loc[:,"Time"], barack_obama.loc[:,"rolling_mean"], color = "#ffa500")
ax4.plot(donald_trump.loc[:,"Time"], donald_trump.loc[:,"rolling_mean"], color = "#00B2EE")

# Title for each graph
ax1.text(x = 12550, y = 1.83 , s = "BUSH", color = "#BF5FFF", weight = "bold")
ax1.text(x = 12210, y = 1.75, s = "(2001-2009)", alpha = 0.3 )

ax2.text(x = 15200,y = 1.83, s = "OBAMA", color = "#ffa500", weight = "bold")
ax2.text(x = 14975, y = 1.75, s = "(2009-2017)", alpha = 0.3)

ax3.text(x = 17800,y = 1.83, s = "TRUMP", color = "#00B2EE", weight = "bold")
ax3.text(x = 17690, y = 1.75, s = "(2017-2021)", alpha = 0.3 )

# Title and subtitle for the whole grid chart
ax1.text(x = 10850, y = 2.15, s = "EURO-USD rate averaged 1.22 under the last three US presidents",weight = "bold",size= 20)
ax1.text(x = 10850, y = 1.97, s = "EURO-USD exchange rates under George W. Bush (2001-2009), Barack Obama (2009-2017), \nand Donald Trump (2017-2021)" )

# Setting x-ticks labels and y-tick labels for the graphs of index one, two, and three
### Removing the x-axis labels and the y-axis labels
for ax in axes:
    ax.set_xticklabels([])
    ax.set_yticklabels([])
### Changes to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],alpha=0.3)
    ax.grid(alpha=0.5) 
### Setting up the x-tick labels
x_ticklabels = ["2001"," ","2003"," ","2005"," ","2007"," ","2009"]
for axes_object in [ax1,ax2,ax3]:
    axes_object.set_xticklabels(x_ticklabels)
# Copyright
ax.text(x = 10690, y = .71, s = "©Bilal Rashid" + " "*104 + "Source: European Central Bank",color = "#f0f0f0", backgroundcolor = "#4d4d4d") 
plt.show()
```

We were able to determine which president of the United States was able to bring the Euro-USD rate to the highest, indicating that the president wasn't the best fit for office, using the graph "EURO-USD rate averaged 1.22 under the last three US presidents." This president's name was George Bush, and he drove the Euro-USD rate to an all-time high because, during his presidency, the United States transitioned from being the world's largest international creditor to the world's largest debtor nation. Under President Bush, the national debt stood at $5.629$ trillion at the end of the calendar year 2000. The federal debt stood at $9.986$ trillion eight years later.
